### 7. Explore whether there’s a connection between age and spending power (income) using a scatterplot

```python
# Reduce previously made sample to only the columns needed
instacart_4 = small [['age','income']]
```

```python
# Create a scatterplot
scatterplot = sns.scatterplot(x = 'age', y = 'income',data = instacart_4)
```

#### The scatter plot here shows that the incomes among the younger age groups particularly ages between 20 to 40 shows the higher concentration at the lower end of the income scale. If we consider the age groups from 40 and above the income status is distributed all over the plot i.e, from bottom to top. From this distrribution we can say that the densest concentration of points for individuals over 40 is significantly higher on the income scale compared to those under 40, which could indicate that the median income increases with age.
